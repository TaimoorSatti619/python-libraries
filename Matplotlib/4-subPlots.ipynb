```python
import matplotlib.pyplot as plt


#plt.subplot(nrows,ncols,index)   1 base indexing

x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1) # 1 row, 2 column, 1st subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2) # 1 row , 2 coloumn, 2nd subplot
plt.bar(x,y)
plt.title('Bar Chart')

plt.show()
```

# advance way

```python
import matplotlib.pyplot as plt

#fig, ax = plt.subplots(nrows, ncols, figsize = (width,height))

fig, ax = plt.subplots(1,2,figsize = (10,5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y,color = 'blue')
ax[0].set_title('Line Plot')

ax[1].bar(x,y,color = 'green')
ax[1].set_title('Bar Chart')

fig.suptitle('Comparison of Line and Bar Charts')

plt.tight_layout()    # full layout   automatic adjustment
plt.show()
```
