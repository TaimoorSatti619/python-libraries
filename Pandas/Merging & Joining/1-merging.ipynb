```python
import pandas as pd
```

```python
# pd.merge(df1,df2,on="Column Name",how="type of join")

df_customers = pd.DataFrame({
    'CustomerId':[1,2,3],
    'Name':['Ali','Hamza','Ahmed']
})

df_orders = pd.DataFrame({
    'CustomerId':[1,2,4],
    'OrderAmount':[250,300,400]
})

```

```python
df_merged = pd.merge(df_customers,df_orders,on="CustomerId",how="inner")
print("Inner Join")
print(df_merged)



df_merged = pd.merge(df_customers,df_orders,on="CustomerId",how="outer")
print("\nOuter Join")
print(df_merged)



df_merged = pd.merge(df_customers,df_orders,on="CustomerId",how="left")
print("\nLeft Join")
print(df_merged)



df_merged = pd.merge(df_customers,df_orders,on="CustomerId",how="right")
print("\nRight Join")
print(df_merged)
```

```python

# 1df = m rows
# 2df = n rows
# m * n rows

df_merged = pd.merge(df_customers,df_orders,how="cross")
print("Cross Join")
print(df_merged)
```
