```python
import pandas as pd
```

```python
data = {
    "Name" : ['Ali','Hamza','Wasi','Zakir','Ahad','John','Kashif'],
    "Age" : [20,23,24,22,25,26,28],
    "Salary" :[50000,70000,80000,52000,53000,55000,58000,],
    "Performance Score" : [85,87,86,80,92,91,93]
}

df = pd.DataFrame(data)
```

```python
avgSalary = df["Salary"].mean()
print(avgSalary)

SumSalary = df["Salary"].sum()
print(SumSalary)

minSalary = df["Salary"].min()
print(minSalary)

maxSalary = df["Salary"].max()
print(maxSalary)
```
