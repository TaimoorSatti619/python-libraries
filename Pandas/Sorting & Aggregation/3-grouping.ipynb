```python
import pandas as pd
```

```python
data = {
    "Name" : ['Ali','Hamza','Ali','Zakir','Ahad','John','Kashif'],
    "Age" : [20,23,24,20,25,24,23],
    "Salary" :[50000,70000,80000,52000,53000,55000,58000,],
    "Performance Score" : [85,87,86,80,92,91,93]
}

df = pd.DataFrame(data)
```

```python
grouped = df.groupby("Age")["Salary"].sum()
print(grouped)
```

## Multiple Columns

```python
grouped = df.groupby(["Name","Age"])["Salary"].sum()
print(grouped)
```
