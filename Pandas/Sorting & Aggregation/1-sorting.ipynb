```python
import pandas as pd
```

## Soring Data 1 Column

```python
# df.sort_values(by="column Name",True/False,inplace = True)

data = {
    "Name" : ['Ali','Hamza','Wasi','Zakir','Ahad','John','Kashif'],
    "Age" : [20,23,24,22,25,26,28],
    "Salary" :[50000,70000,80000,52000,53000,55000,58000,],
    "Performance Score" : [85,87,86,80,92,91,93]
}

df = pd.DataFrame(data)

df.sort_values(by="Age",ascending=False,inplace=True)
print("Sorted Age by Descending ")
print(df)

df.sort_values(by="Age",ascending=True,inplace=True)
print("Sorted Age by Ascending ")
print(df)
```

## Sorting multiple columns

```python
df.sort_values(by=['Age','Salary'],ascending=[True,False],inplace=True)
print("Sorted Age and Salary by Descending ")
print(df)
```
