```python
import pandas as pd
```

```python
data = {
    "Name" : ['Ali','Ahmed','Hamza','Wasi','Zakir','Ahad','John','Kashif'],
    "Age" : [20,21,23,24,22,25,26,28],
    "Salary" :[50000,60000,70000,80000,52000,53000,55000,58000,],
    "Performance Score" : [85,90,87,86,80,92,91,93]
}

df = pd.DataFrame(data)
```

## Updating  value of column

```python
# df.loc[rowIndex,"Column name"] = newValue
df.loc[0,"Salary"] = 55000
print(df)
```

```python
# Increaing Salary by 5%

df["Salary"] = df["Salary"] * 0.5
print(df)
```
