```python
import pandas as pd
```

```python
data = {
    "Name" : ['Ali','Ahmed','Hamza','Wasi','Zakir','Ahad','John','Kashif'],
    "Age" : [20,21,23,24,22,25,26,28],
    "Salary" :[50000,60000,70000,80000,52000,53000,55000,58000,],
    "Performance Score" : [85,90,87,86,80,92,91,93]
}

df = pd.DataFrame(data)
```

## Removing Column

```python
# df.drop(Columns = ["columnName"],inplace=True)   inplace mean change in original data
df.drop(columns=["Performance Score"],inplace=True)
print(df)
```

```python
##  Deleting Multiple Columns
# df.drop(Columns = ["columnName","columnName"],inplace=True)   inplace mean change in original data
df.drop(columns=["Salary","Age"],inplace=True)
print(df)
```
