```python
import pandas as pd
```

```python
data = {
    "Name" : ['Ali',"Ahmed",'Hamza','Wasi','Zakir','Ahad','John','Kashif'],
    "Age" : [20,None,23,24,22,25,26,28],
    "Salary" :[50000,None,70000,80000,52000,53000,55000,58000,],
    "Performance Score" : [85,None,87,86,80,92,91,93]
}

df = pd.DataFrame(data)
```

```python
# linear  , polynomial, time
df.interpolate(method="linear",axis=0,inplace=True)
print(df)
```

## Linear

```python
data = {
    "Time" : [1,2,3,4,5],
    "Value" : [10,None,30,None,50]
}

df = pd.DataFrame(data)
print("Before interpolation")
print(df)

df['Value'] = df['Value'].interpolate(method="linear")
print("After interpolation")
print(df)
```
