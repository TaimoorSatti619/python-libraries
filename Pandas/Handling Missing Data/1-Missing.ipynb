```python
import pandas as pd
```

```python
data = {
    "Name" : ['Ali',None,'Hamza','Wasi','Zakir','Ahad','John','Kashif'],
    "Age" : [20,None,23,24,22,25,26,28],
    "Salary" :[50000,None,70000,80000,52000,53000,55000,58000,],
    "Performance Score" : [85,None,87,86,80,92,91,93]
}

df = pd.DataFrame(data)
```

```python
print(df.isnull())
print(df.isnull().sum())
```

## Handling Missing Value


```python
# axis 0   = rows
# axis 1   = columns
# dropna(index,inplace=True)

print(df)
df.dropna(inplace=True)
print(df)
```

```python
# Method 2  
# fillna(value,inplace=True)



# using mean of age
#df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Age'] = df['Age'].fillna(df['Age'].mean())

print(df)
```
