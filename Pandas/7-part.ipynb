```python
import pandas as pd
```

```python
data = {
    "Name" : ['Ali','Ahmed','Hamza','Wasi','Zakir','Ahad','John','Kashif'],
    "Age" : [20,21,23,24,22,25,26,28],
    "Salary" :[50000,60000,70000,80000,52000,53000,55000,58000,],
    "Performance Score" : [85,90,87,86,80,92,91,93]
}

df = pd.DataFrame(data)

print("Sample DataFrame")
print(df)
```

```python
print("Names (single column returns series)")
name = df["Name"]
print(name)


# selecting multiple columns
subset = df[["Name","Salary"]]
print("\nSubset with Name and Salary")
print(subset)
```

## Rows filter

```python
high_salary = df[df["Salary"] > 55000]
print("Employees with Salary > 55000")
print(high_salary)
```

```python
filtered = df[(df["Salary"] > 55000) & (df["Age"] > 23)]
print("Employees list Age > 23 and Salary > 55000")
print(filtered)
```

```python
# USing OR Condition

filtered = df[(df["Performance Score"] > 90) | (df["Age"] > 24)]
print("Employees list Age > 23 OR Salary > 55000")
print(filtered)
```
