```python
import pandas as pd
```

```python
data = {
    "Name": ['Ali','Ahmed','Hamza'],
    "Age": [10,20,30],
    "City": ['rwp','isb','lhr']
}

df = pd.DataFrame(data)
print(df)
```

```python
#  save CSV  file
df.to_csv("output.csv",index=False)
```

```python
#  save EXCEL  file
df.to_excel("output.xlsx",index=False)
```

```python
#  save JSON  file
df.to_json("output.json",index=False)
```
